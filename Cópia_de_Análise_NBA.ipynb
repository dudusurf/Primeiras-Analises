# POR JOGADOR
REBOTES/ASSISTÊNCIAS/PONTOS/TOTAL

```python
nba_jogador = pd.read_excel('/df_nba.xlsx', index_col= 0 )
nba_jogador.fillna(0, inplace= True)
nba_jogador = nba_jogador[['Tm','TRB','AST','PTS','3P']]
for index, row in nba_jogador.iterrows():
    nba_jogador.loc[index, 'TOTAL'] = row['PTS'] + row['AST'] + row['TRB']
    nba_jogador.loc[index, 'PTS/AST'] = row['PTS'] + row['AST']
    nba_jogador.loc[index, 'PTS/TRB'] = row['PTS'] + row['TRB']
    nba_jogador.loc[index, 'TRB/AST'] = row['TRB'] + row['AST']

```
